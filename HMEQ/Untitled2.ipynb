```python
Donnee_mis=hmeq.isna().sum()
Prop_mis=(Donnee_mis/hmeq.shape[0])*100

PD=pd.DataFrame({
    #'variable':hmeq.columns,
    'Number of missing data' : Donnee_mis,
    'Prop of missing data' : Prop_mis
})
PD
```

```python
hmeq_quan=hmeq.select_dtypes(include=['int','float'])
hmeq_qual=hmeq.select_dtypes(include='object')
hmeq_quan.shape
```

```python
L_inf=[]
L_sup=[]

for var in hmeq_quan.columns[1:]:
    if Prop_mis[var]<=5:
        L_inf.append(var)
    else :
        L_sup.append(var)
```

### Imputation par la moyenne ou la médiane

```python
New_hmeq=pd.DataFrame({})
for var in L_inf:
    M=hmeq[var].median()
    New_hmeq[var]=hmeq[var].fillna(M)
    
New_hmeq.head(5)
```

### Imputation par KNN

```python
from sklearn.impute import KNNImputer
```

```python
imputer= KNNImputer(n_neighbors=3)
```

```python
for var in L_sup:
    New_hmeq[var]=imputer.fit_transform(np.array(hmeq[var]).reshape(-1, 1))
    
    
New_hmeq.head(5)
```

```python
from statistics import mode
for var in hmeq_qual.columns:
    Mod=mode(hmeq[var])
    New_hmeq[var]=hmeq[var].fillna(Mod)
    
New_hmeq.head(5)
```
